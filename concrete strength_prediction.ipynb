```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
import pickle
```

```python
conc= pd.read_csv('Concrete_Data.csv')
```

```python
conc.head(4)
```

```python
conc.describe
```

```python
conc.isnull().any()
```

```python
X = conc.drop('Strength', axis =1)
y = conc['Strength']
```

```python
X.columns
```

```python
model_rf = RandomForestRegressor(random_state=24)
```

```python
model_rf.get_params()
```

```python
kfold = KFold(n_splits=5, random_state=24,shuffle=True)
parameters = {'max_features': np.arange(1,7)}
```

```python
model = GridSearchCV(model_rf, param_grid=parameters,cv=kfold,verbose=3)
```

```python
model.fit( X , y )
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
best_model = model.best_estimator_
```

```python
best_model.predict([[540.0,	0.0,	0.0,	162.0,	2.5,	1040.0	,676.0,	28]])
```

```python
file_pkl = open("concrete.pkl", 'wb') 
pickle.dump(best_model, file=file_pkl)
```
